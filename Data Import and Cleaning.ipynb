```python
# Creating a dataframe with the columns that I want to keep

clean_bom_df = bom_df[['title', 'domestic_gross']]
display(clean_bom_df.head())
clean_bom_df.info()
```

```python
# Creating a dataframe with the columns that I want to keep

clean_tn_movies_budgets_df = tn_movies_budgets_df[['movie', 'production_budget', 'domestic_gross']]
display(clean_tn_movies_budgets_df.head())
clean_tn_movies_budgets_df.info()
```

```python
# verify similarities between bom_df and tn_movies_budgets_df as far as movies represented go
pd.value_counts(clean_bom_df['title'].isin(clean_tn_movies_budgets_df['movie']))
```

```python
# Here I reduced the numbers to be representing millions, and rounded for uniformity. I also dropped the excess columns

clean_bom_df['domestic_gross_in_mil'] = clean_bom_df['domestic_gross'].apply(lambda x: x/1000000)
clean_tn_movies_budgets_df['title'] = clean_tn_movies_budgets_df['movie']
clean_tn_movies_budgets_df['domestic_gross_in_mil'] = clean_tn_movies_budgets_df[
    'domestic_gross'].apply(lambda x: x.replace(",", "").replace("$", ""))
clean_tn_movies_budgets_df['domestic_gross_in_mil'] = clean_tn_movies_budgets_df[
    'domestic_gross_in_mil'].apply(lambda x: round(int(x)/1000000, 1))
clean_tn_movies_budgets_df['production_budget_in_mil'] = clean_tn_movies_budgets_df[
    'production_budget'].apply(lambda x: x.replace(",", "").replace("$", ""))
clean_tn_movies_budgets_df['production_budget_in_mil'] = clean_tn_movies_budgets_df[
    'production_budget_in_mil'].apply(lambda x: round(int(x)/1000000, 1))

clean_bom_df.drop('domestic_gross', axis = 1, inplace = True)
clean_tn_movies_budgets_df.drop(['movie',
                                 'production_budget',
                                 'domestic_gross'], axis = 1, inplace = True)

display(clean_bom_df.head())
clean_tn_movies_budgets_df.head()
```

```python
# Merging the above two DFs so that movies aren't represented more than once, then checking for any duplicates that
# may have slipped through

gross_df = pd.merge(clean_bom_df, clean_tn_movies_budgets_df, how = 'outer', on = 'title')
display(gross_df.head())
gross_df['title'].value_counts()
```

```python
# Before dealing with the remaining duplicates, I wanted to condense the two seperate 'domestic gross' columns

gross_df['domestic_gross_in_mil_x'].fillna(gross_df['domestic_gross_in_mil_y'], inplace=True)
gross_df['dom_gross_mil'] = gross_df['domestic_gross_in_mil_x']
gross_df.drop(['domestic_gross_in_mil_x', 'domestic_gross_in_mil_y'], axis = 1, inplace = True)
gross_df.head()
```

```python
#See how many of these duplicates are the same across all columns, rather than just name

gross_df.duplicated().value_counts()
```

Here I realized that if I wanted to deal with movies of the same name (such as remakes) that it would have been best to keep the years that each movie was made. However, since I don't require the release date of a movie to make statistical assumptions about the profitability of a movie, and since there is a single absolute duplicate, all I need to do with this data frame is to drop the one 'True' row.

```python
# Conducting the required drop, then making sure it worked
gross_df = gross_df.drop_duplicates()
gross_df.duplicated().value_counts()
```

```python
# Because all the values in 'title_rating_df' I can execute a left join to get , then verify that
# each movie is only represented once
ratings_df = pd.merge(clean_imdb_title_rating_df, clean_imdb_title_basic_df, how='left', on='tconst')
display(ratings_df.head())
ratings_df['tconst'].value_counts()
```

```python
full_ratings_df = pd.merge(clean_tmdb_movies_df, ratings_df, how='outer', on = 'title')
full_ratings_df.head()
```

```python
full_ratings_df = full_ratings_df.drop_duplicates()
full_ratings_df.duplicated().value_counts()
```

```python
full_ratings_df.drop('release_date', axis=1, inplace = True)
```

```python
full_ratings_df.drop('tconst', axis=1, inplace = True)
full_ratings_df.drop('runtime_minutes', axis=1, inplace = True)
```

```python
full_ratings_df.head()
```

```python
# Dealing with null values by turning vote counts to zero, and setting average ratings
full_ratings_df['averagerating'].fillna(full_ratings_df['vote_average'], inplace = True)
full_ratings_df['vote_average'].fillna(full_ratings_df['averagerating'], inplace = True)
full_ratings_df['vote_count'].fillna(0, inplace = True)
full_ratings_df['numvotes'].fillna(0, inplace = True)
print(full_ratings_df['averagerating'].value_counts)
print(full_ratings_df['vote_average'].value_counts)
print(full_ratings_df['vote_count'].value_counts)
print(full_ratings_df['numvotes'].value_counts)
full_ratings_df.head()
```

```python
# Condensing the final ratings DataFrame
full_ratings_df['avg_rating'] = (full_ratings_df['vote_average'] + full_ratings_df['averagerating'])/2
full_ratings_df['total_votes'] = full_ratings_df['vote_count'] + full_ratings_df['numvotes']
full_ratings_df.drop(['vote_average', 'vote_count', 'averagerating', 'numvotes'], axis = 1, inplace = True)
full_ratings_df.head()
```
